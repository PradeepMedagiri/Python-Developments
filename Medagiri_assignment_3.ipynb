```python
def function():
    import random
    keys=[]
    values=[]
    choose='yes'
    try:
        x=input("Give the name of the “words and their meanings” file:")
        with open(x,encoding="ISO-8859-1") as z:
            for line in z:
                present = line.split(",")
                keys.append(present[0])
                values.append(present[1])
        while(choose=='yes'):
            a=random.choice(keys)
            b=keys.index(a)
            c=values[b]
            scramble=''.join(random.sample(a,len(a)))
            print("Unscramble the following letters to form a word. Type “?” for the meaning of the unscrambled word: ",scramble)
            fanswer=input("Enter the answer [or ? for the meaning]: ")
            while(fanswer!=a):
                if(fanswer=='?'):
                    print('The word means: ',c)
                else:
                    print('Wrong, try again')
                fanswer=input("Enter the answer [or ? for the meaning]: ")
            print('You got it!')
            choose=input('Do you want to continue [yes or no]: ')
        if(choose=='no'):
            print("Goodbye!")
        else:
            print("Invalid input: Please run/try again")
    except:
        print("Please enter valid filename/ check for specific file format")
        function()
function()
```
